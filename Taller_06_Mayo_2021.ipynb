```python

import re
import pandas as pd 
from pprint import pprint

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import pyLDAvis.gensim_models
from gensim.models import LdaModel 
from gensim.corpora import Dictionary
from pprint import pprint 


import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
data = pd.read_csv("C:/Users/usuario/Desktop/NLP/22_Febrero/reviews_vidjew_es.csv")
data.head()
```

## [12 pts] Punto 1: Hacer pre-procesamiento del texto

```python
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = [palabra for palabra in texto.split() if palabra not in stopwords]
    return texto
 
data['Preprocesado'] = data['review_body'].apply(lambda texto: pre_procesado(texto))
 
data.head()
```

## [10 pts] Punto 2: Modelo de LDA

### crear una representación de los documentos en formas de diccionario 

```python

dictionary = Dictionary(data['Preprocesado'].values)
dictionary
```

###  Filtrar palabras frecuentes o infrecuentes 

```python
dictionary.filter_extremes(no_below=5, no_above=0.5)
```

###  Corpus

```python
corpus = [dictionary.doc2bow(text) for text in data['Preprocesado'].values]
```

###  Entrenar el modelo

```python

model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=50)
model.print_topics(num_words=8)
```


## [25 pts] Punto 3: Visualización de LDA

```python
import pyLDAvis.gensim_models
from gensim.models import LdaModel 
from gensim.corpora import Dictionary
from pprint import pprint 
```

```python
lda_display = pyLDAvis.gensim_models.prepare(model, corpus, dictionary)
pyLDAvis.display(lda_display)
```
